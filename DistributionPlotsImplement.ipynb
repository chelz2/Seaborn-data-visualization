# Distribution Plots in Seaborn --> implemention


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
df = pd.read_csv("dm_office_sales.csv") # --> readind in tabular data csv as pandas df

df.head()
```

```python
# checking distribution of salary in this datasets

#RUG Plot ---> most basic distribution plot

sns.rugplot(x='salary', data=df)  #---> as we dont have meaning y-axix, so only pass in datasource & x-axis

#----> we get tick?line for every single salary 
```

```python
# height param ---> to invrease the height of tick in rug plot

plt.figure(figsize=(4, 3), dpi = 150) #---> customize the figue object if we want 

sns.rugplot(x='salary', data=df, height=0.5)  #---> we pass in the height to be a percentage of y-axis, 50% of y-axis
```

###### NB ---> Y -axis values are not menaingful iher ein Rug plotd


###### ------> Nb- sometimes peaple think of rug plot as one - dimensional sctter plots, since it( Rug plot) is a scatter , *BUT it is happeng in 1 diminsional so we only have ticks here

```python
# Histogram Plots ---> .displot() & histplot


# NB --> in latest seaborn libray there is both .distplot() which is depracated, so we ll have to use .displot() method

#----> NB- i.e seaborn now has two function .histplot() for histogram & other one is .displot() which we ll have to use 
```

```python
# Distribution plot --> Histogram 

sns.displot(data=df, x='salary') #---> plotting distribution plot of salary
```

```python
# bins param ---> we can pass in no. of bins we want only

sns.displot(x='salary', data=df, bins=10)  # ----> creates a histogram plots of salary with bin size equals to 10  

#-----> we can just see general trend in salary
```

```python
sns.displot(x='salary', data=df, bins=100) #---> breaking up x-axis into 100 qual size bins histogram plot

# ---> we can now get too much detail of the salary distribution
 
```

###### ----> i.e this bin size is the most common feature to edit in histogram plots

###### ---> NB - we can also add in other params like *Grids & *styles for histogram since with default white background it is hard to match up the height of histogram column with y-axis

```python
# adding grid in histogram plot --> which we can pass above the .distplot() method call

# sns.set(style = 'whitegrid') # ---> there are diff grid style we can pass (refer docs)

sns.displot(x='salary', data=df, bins = 30)
```

###### NB ---> i.e once we se pass in the style background, it set it for the rest of the notebook, it is a high level command

###### -----> if we dont like the style we can go to notebook kernel  & run shutdown & run all button with .set style method commented out which ll reset style  to default

```python
# NB ---> all the aguments we learnt in MAtplolib customiztion --> to edit color, linewidth etc is available in .displot() 
        # method call


# shift + tab ---> we'll see all params & docs availble tou in .displot() method & ALSO **Kwarg which refers to all kwargs 
                # available to us in matplotlib 

sns.displot(data=df, x='salary', color= 'orange', edgecolor='blue', linewidth=3) #--> adding color, lw etc
```

```python
# Distibution Plot ---> .histplot() 

sns.histplot(data=df, x='salary') #----> similar to .displot(), creates a histogram
```

###### NB ---> i.e it is upto us to decide what we want to use between .displot() or >histplot() for histogram

###### ----> recommended we use .displot()  which is more gereralize version of this histogram plot which just show us the distribution

```python
# Kde param ---> we can see KDE plot on top the .distplot() histogram

# rug param ---> we can also paas in rug param to see rug plot

sns.displot(data=df, x='salary', kde=True, rug=True) # passing in kde param show kde plot & rug param to see rug plot along with histogram

#----> i.e. we can see all the distrubution type plot using .displot() plot
```

###### NB -----> i.e this .displot() method is the higher level plot for distribution that can show all three option if we want

###### -----> by default it is a histogram plot ****but we can also pass in kde & rug params To see all three distribution plot type

```python
# KDE plot 

sns.kdeplot(data=df, x="salary")  #----> plots out KDE curve for continous feature salary cloumn 
```

```python
# --> building KDE plot starting with Rug plot, then with Histogram & then with the KDE

# Age dataset example

np.random.seed(42)
sample_age = np.random.randint(0,100,200)  #---> creates 200 random integer sample age dataset between year 1 to 100

sample_age
```

```python

sample_ages = pd.DataFrame(sample_age, columns=['age']) # ---> create pandas dataframe with age column base of sample_age array

sample_ages  # dataframe with age column
```

```python
sns.rugplot(data=sample_ages, x='age') # ---> Rug plot to see the distribution of age
```

```python
sns.displot(data=sample_ages, x='age', rug=True, bins=20) #---> displot to see histogram & rugplot to see distribution of data 
                                                                # passing in other params like binsize etc
    
# We cans see rug plot above the histogram plot of the age 
```

```python
# NB ---> the reason we want to edit Bin size is because ther is anorther analogous arg/param which is editing bandwidth for 
        # the kernel of kde plot
    

sns.displot(data=sample_ages, x = 'age', bins=30, rug=True, kde=True) # --> pass in kde to better understand hist & rug plot 
```

###### ----> we can see kde plot is cut off at 0 & 100 as there is no age below 0 or above 100 in sample age dataset


###### ---> *but in some cases we may have actual value below or above the limits in datasets


```python
sns.kdeplot(data=sample_ages, x = 'age')  # ---> we can see by default when these gauasian plot above each ticks together as 
                                            # as the Kde plot , we' ll see probality for values outside of range in datasets
```

```python
#clip param in kde plot ----> we can cut off the kde plot like in displot by passing in clip param

sns.kdeplot(data=sample_ages, x='age', clip=[0,100]) # ---> cut off KDE plot at 0 & 100
```

###### NB ---> Y- scale in kde plot is a density which presents the probabilty of picking that particular instance in datasets

###### -----> *AND taking the area in between ranges will give us probability of range in dataset




##### ------> we can also edit bandwidth passing banwidth param --> bw_adjust
        
         # if pass in bandwidth as very smaall value ---> then with very small bandwidth, we ll have a very skinny 
           gauassian distribution stack on top of each of these rug plots
                
                # ---> it'll pick up lot variance ?noise here
                
          # AND as pick high value, we pick up a less & less of the noise
          
          
         # *BUT --> if we pass bandwidth too width, it ll then eventually pick nothing of the signal & we'll get curve that 
                     look like normal distrubution which may not be what are going for
                     
            

```python
#Bandwidth param

sns.kdeplot(data=sample_ages, x='age', clip=[0,100], bw_adjust=0.02)  # ----> paasing in very small bandwidth

# ---> picks lot of variance / noise 
```

```python

sns.kdeplot(data=sample_ages, x='age', clip=[0,100], bw_adjust=0.2) # ---> makes jaggery as it still picks more of variance
```

```python

sns.kdeplot(data=sample_ages, x='age', clip=[0,100], bw_adjust=0.5)  # ---> picks less noise as we make bandwidth more wider
```

```python

sns.kdeplot(data=sample_ages, x='age', clip=[0,100], bw_adjust=3) # ---> picks nothing of the signal & ll get big normal dist curve
```

```python
# NB ----> i.e in reality, dataset distribution (kde plot) should like more even / uniform  & not like normal dist curve


#NB ----> we should adjust bandwidth size reasonably & not pick too small size

# ----> i.e we may manually want adjust bandwidth to show little more behaviour the data distribution or may just want to keep 
       # keep the default kDE plot 
    
    
sns.displot(data=sample_ages, x='age', rug=True, kde=True)  # ----> default KDE plot
```

```python
sns.kdeplot(data=sample_ages, x='age', clip=[0,100], bw_adjust=0.6, fill=True) # ---> passing in reasonable bandwidth

# we can also shade the curve area passing in 'fill' param ---> 'shade' is depracated
```
