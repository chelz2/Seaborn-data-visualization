# Grid Plots in Seaborn

----> analogous to (we can think it as) matplotlib's subplots

----> automatically creates grid for subplots baed off a categorical column

----> Recap: in matplotlib we had to pass in specific no. of columns or rows for subplots 

---->*BUT, in seaborn, we can just pass in the name of the column that we want seaborn to automatically map to the subplot's grid

----> i.e many of these seaborns builtin calls runs on top of this grid system, allowing us to simply define extra parameter calls for rows or columns 

----> also, directly calling grid system also allows us to heavily customize plots, specifically pairgrids that pair plot runs on top of

####  * two main type of grid plots in seaborn grid syatem :-
        
        - category plot
        - pair plot

##### Category plot in Grid plot system

```python
# catplot ----> statistical distribution within categoried that we can explore

# ----> takes in additional param of rows & column which'll actually define this subplot grid  that is created below yhis 

#-----> & then we just have to say waht kind of plot we want passing in kind param, eg - violin plot, boxlot etc 


sns.catplot(data=df, x='gender', y='math score', kind='box') #NB -->i.e when we call boxplot, it actually does this catplot
                                                                # call underneath where it is specifying underneath kind type to 
                                                                 # to be box
                                                                    
                                                                                            
```

### PairGrid

```python
# pair grid

sns.pairplot(data=df)  # ----> recap- takes all numerical features & automaticlly creates a pair plot comparisorn between them
```

----> we have scatter plots & histogram on diagonals


----> NB - i.e, what is happenning underneath is .pairplot() calls the pair grid system functionality which creates this grids of axes & fills it in with plots


---> i.e, we can explore the pair grid system

```python
# PairGRid method call

sns.PairGrid(data=df)  # ---> calling in PairGrid
```

##### ----> creates the grid axes automatically for us based of columns in dataframe or columns we pass in 

----> i.e, what it waits for is the actual mapping command that it wants us to provide 

----> we set it equal to a variable & then we start calling mapping command methods (3 method)

     - upper mapping :  for plots in upper half of the diagonal in grid
        
     - diagonal mapping - for plots in diagonal of grid
    
     - lower mappig - for lower half of the diagonal in grid

```python
# plots for upper half, diagonal, & lower half in Pair Grid axes

g = sns.PairGrid(data=df) # ----> creates the undrrelying empty pair grids for all contionuos feature columns in dtaframe

g = g.map_upper(sns.scatterplot)  # ---> passing in sactter plot types for upper half for two contnuos feature plot

g= g.map_diag(sns.kdeplot)  # passing in kde plot for diagonal for single feature plot, eg- math v/s math etc

g = g.map_lower(sns.kdeplot) # ----> passing in kde plot for lowewr half for two continuos feature plot
```

##### ----> we can see scatter plots in upper half, kde plot in lower half, & kde plot or histogram plot in diagonal 
    

    #NB --> i.e, the mapping should make sense for the plot type
    
    #------> .ie upeer half & lower half  mapping plot should a plot type tthat can take two continuos features & compare them in X & y axes 
    
    #-----> i.e for diagonal, we should pass a plot types that makes sense for single feature eg- histplot, kde plot etc for dta   distrubutin 
    
      Eg - scatter plot showing relation between two continuos feature 
      
      
#---> this is the way we can customize the pair plots to whatever we want 


#---> It (PairGrid) is not common to do deeper dive on pairgrid, i.e calling pairplot is fine & then calling joint plot to further explore data to compare betwwen diff continuos feature is fine

( but if we want for some final visualization, then we can create this PairGrid axes & then amnually map the upper, lower diagonal to whatever reasonable plot types we want to display)


```python
# params in PairGrid method ---> we can pass in other param calls on all the plots in grid, eg - hue param

g = sns.PairGrid(data=df, hue='gender')  # passing in hue param as gender category 

g = g.map_upper(sns.scatterplot)  

g= g.map_diag(sns.kdeplot)  

g = g.map_lower(sns.kdeplot) 

g = g.add_legend() # ---> * have to manually add legend for hue category in pair grid plot
```

---> we can see it plots separetly based of hue category gender - male & female

#---> NB- we have to do lot more than just call pairplot() method *But we can lot more customization in PairGrid() plot call


### Facet Grid 

######    Ctaplot()
- the catplot which we seperate out by column and row is technically running off another seaborn grid object called SNS FacetGrid
              




-----> # NB - techically we never use this, since it has lot manual to create exact same default plot created by catplo() method

```python
g = sns.FacetGrid(data=df,col='gender', row='lunch')  # ---> creates facet grid object 

g= g.map(plt.scatter, 'math score', 'reading score', edgecolor='w') #--> using map calls to map thing to this facet 
                                                                            # grid 
g.add_legend()
```
