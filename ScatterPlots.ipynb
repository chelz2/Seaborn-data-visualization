# Scatter plots in Seaborn
------> Shows retionship  between two ***continuos features 

```python
#Continuos features ---> numeric variables that can take any value between any two values

#------> the features we can continue to divide into smaller & smaller chunks

#Eg ---> Age 28 can be said as by more smaller chunks like 28 years 2months 1 week etc

#----> we can say same for height, salary, temperature, prices etc (continiously divide into smaler chunks)
```

```python
# ----> i.e, it(continius feature) allows for a value to always be between two values

# continuos feature V/S categorical features ----> *NB - i.e continuos should not be confused with categorical features 

# -------> **categorical features represents unqiue & distnct unqiue categories (i.e where there is no in between values somtimes)
#--------> eg - distinc colors (red, green, blue), Shapes , Names etc
```

```python
# scatter plots & seaborn

#----> line up two continiuos features & then plot them out as coordib=nate x vs y

# Example --> explore relation between employee salaries & sales amount they make selling certain items

# Coloring & styling -----> we can also coloring & styling in scatter plot based of other columns in df table

#----> coloring based off experience, i.e we can also relation see realtion betwwen salry & experience other than sales by coloring scatter plot

#---> ALso, .i.e we can have coloring & styling based of categorical columns(props), eg- based of category of education- high school, bachelor, master etc
```

```python
# NB ---> we dont want to be carried away by plotting as much information in a plot

# eg --> for relation between salary & sales, we can just show scatter plots between the two 

```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns     # recommende we import seaborn as * sns
```

```python
df = pd.read_csv('dm_office_sales.csv') # ----> graving tabular datatable pandas dataframe 

df.head()
```

```python
#Nb ---> for relation between salary & sales, i.e we can answer it numerically by taking coefficient correaltion between salary & sales

# Or ----> we can just plot out the relations

sns.scatterplot(x='salary', y='sales', data = df)  # we pass in dataframe & define x-y axis with columns we want to plot  
```

##### Refer ---> Seaborn docs for more

#----> Seaborn is creating this plot using matplot, which means we can grab this figure object canvas & start editing it before we even call scatter plot method with seaborn 

#NB --->custom figure object canvas &  scatter plot method by seaborn should be in same cell

#---> *NB :- Seaborn just calling matplotlib in the backend

```python
plt.figure(figsize=(12,5), dpi=150)  # custom sizing the figure object canvas using matplot


sns.scatterplot(x='salary', y='sales', data=df)  # ----> seaborn scatter plots on customize figure object canvas  
```

```python
# Seaborn parameters --> why & when to use & specific scatter plot parameters

#hue param --> takes in column & color all these dots in scatter plot based on column(i.e default scatter plot colors all dots as same) 

plt.figure(figsize=(12,6), dpi=150)

sns.scatterplot(x='salary', y='sales', data=df, hue='level of education') #---> hue tells to color by the clumn we pass

# & automatiically creates a Legends for us based of this particular categorical column ( level of education)
```

```python
# paasing in continous-dtype column as hue --->  eg - passing in salary as hue ( doubling down on info )

plt.figure(figsize=(12,5), dpi=150)

sns.scatterplot(x='salary', y='sales', data=df, hue = 'salary') # passing in salary both as x axis & hue (doubling down on info )
```

#####  ---> if we choose a continious column as hue, then it creates a gradient of a color in scatter plots & it automatically choose some major stoping pts to show us  different color code

#### NB- i.e we can hue by continiuos or categorical column, which means using 2D plots (X - Y coordinate ) we can show  3 dimension of infos

eg --> salary is 1D, sales is 2nd D & level of edcation is 3rd D ( i.e,  3dimension infos in 2D plot ! ) 

```python
# Choosing colors for hue ----> we have to choose pallete from mataplotlib's color map (refer docs)

#recommded --> choose perceptually uniform sequential colormaps ( so even color blind peaple can see the diff in color in plot)

#based of column type (cotinious or categorical) , we chose colormaps

#---> eg - for categorical type, we choose categorical colormap which is call qualitative colormap in matplot
```

```python
# colormap in scatter plot (seaborn) -----> we pass in pallete parameter

plt.figure(figsize=(12,6), dpi=150)


sns.scatterplot(x='salary', y='sales', data=df,hue='level of education', palette='Dark2')
```

```python
# Size param ---> sizing dots in scatter plot

# we can also size scatter plot dots based off another column ----> eg- size of dots based off work experience

plt.figure(figsize=(12,6))

sns.scatterplot(x='salary', y='sales', data=df, size='work experience') #---> showing 3D infos in 2D plot 
 
```

###### NB ---> we can size the dots base off any column type, i.e it dont have to be continous type *BUT should make sense for something as continious
eg- work experience column

###### NB --> we can also both size & color 


```python
# s parameter -- > general size for all dots/markers in plot

plt.figure(figsize=(12,6))

sns.scatterplot(x='salary', y='sales', data=df, s=200) #----> increases all the dots by 200 pts
```

```python
# alpha parameter --> passing in transparency in dots in scatter plots ---> 0(fully transparent) & 1 (fully opaque by default)

plt.figure(figsize=(12,6))

sns.scatterplot(x='salary', y='sales', data=df, s=200, alpha = 0.2 ) # ---> making markers transparent by 80%
```

###### ----> this way we can clearly see where points are most stack over each other 

eg -- we can see most salaries are between 60k to 120k

```python
# style parameter -----> that is markers we are using for data pts , eg - dots , plus , triangle etc

plt.figure(figsize=(12,6))

sns.scatterplot(x='salary', y='sales', data=df, s=200, style='level of education',
               hue='level of education') 
# ---> automatically style dots to diff shapes & colors base of education level category, passing in multiple params


plt.savefig('scatterplot.jpg') # -----> same as matplot, saving this plot as scatterplot.jpg image file
```

###### ----> we dont show more infos by passing in hue & style, *BUT makes it easier to read the plot


###### NB --->  *Style parameter is helpful if publishing the plot in black & white color, and  also *Style should be categorical feature type column when assingning since we are categorizing things (datas) base of shapes 


###### ----> we can keep adding different params on & if we want we can also size by another columns ( we can do all things available to us in seaborn)


###### ----> NB - i.e the main thing is we define the X & Y co-ordinate and the datasource( dataframe )  ** AFTER which everything is styling & customization
