### box plot 


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv') # readind in csv tabular data as pandas dataframe table

df.head()  # every rows represnt a student & their score & other ategorical info like race, course, etc
```

```python
# we can cexplore understanding distuobution across the continuos feature of score & how it relate to diff categories

# NB - if we can see the distubution about everything, eg -distubution of all math score 

sns.boxplot(data=df, y='math score') # ---> boxplot distribution for all of math score not seperated by any category
```

```python
sns.boxplot(data=df, y='reading score', x='test preparation course') #---> boxplot distubution for math score 
                                                                        # base of test prep course category

# showimg boxplot distrubution of math score base of wheter student took test prep or not   
```

```python
sns.boxplot(data=df, y='math score', x='test preparation course') 

#---> boxplot distubution for reading score against the test preparation 
```

---> we can see test preparation help score better than not preparing for test 

---> with all the boxplot feature like whisker, median & outliers etc

```python
plt.figure(figsize=(12,5))

sns.boxplot(data=df, y='math score', x='parental level of education')

# ---> boxplot score against parent's level of education
```

--> we can see parents level of education does help children to do better

```python
# we can also do subcategories using *hue param based of another categorical feauter

plt.figure(figsize=(12,5))

sns.boxplot(data=df, y='math score', x='parental level of education',
            hue='test preparation course') # sub categorize within larger category in boxplot


plt.legend(bbox_to_anchor=(1.05, 0.5))

# showing score against parent level of education & hue subcategory showing if student prepared for test
```

---> we can student who prepared for test did well regardless of thier parent's educatiom 


```python
# sshowing categories on y-axis ---> interchange y axis for categorical feature & vice versa

plt.figure(figsize=(5,8))

sns.boxplot(data=df, y='parental level of education', x='math score',
            hue='test preparation course', palette='Set2')  # passing categorical feature (level of edu) on y-axis

plt.legend(bbox_to_anchor=(1.05, 0.5))
```

---> box plot, we spicify a continuos feature against a categorical one & if we want we can sub-category base of hue coloring

--- & also we can also edit default coloring passing in pallete param with matplot's colormap we want

### Violin Plot

```python
# violin plot ---> we can also have violin plot instead of boxplot if we want

# ----> shows the mirroed KDE plot of continios feature data pts

sns.violinplot(data=df, x='reading score', y='parental level of education')  

# violin plot distributon of score against parent level of education same as boxplot  
```

---> i.e what this violin plot is showing is KDE plot with a mini boxplot at center for parental level of edu against reading score

```python
# like boxplot, we can pass in hue param to sub categorize

plt.figure(figsize=(5,8))

sns.violinplot(data=df, y='parental level of education', x='math score',
            hue='test preparation course', palette='Set2')   

plt.legend(bbox_to_anchor=(1.05, 0.5))
```

--> we can see violin plot right next to each other base of hue seperation of subcategorise test preparation

-->* recommended we use boxplot instead of violinplot if we want to have a hue seperation for sub category

--> also we can see violin plot touch each other & is not easy to read. Also most peaple are not familiar with violinplot

```python
# split Param ---> makes 1 violin plot base of top half of one catgory & bottom half of another

plt.figure(figsize=(5,8))

sns.violinplot(data=df, y='parental level of education', x='math score',
            hue='test preparation course', palette='Set2', split=True)  

plt.legend(bbox_to_anchor=(1.05, 0.5))
```

---> we now have one half of category who completed test preparation (lower half)  & other half with category who didnot

```python
# inner param ---> we can get rid of inner boxplot in violinplot if we want to

plt.figure(figsize=(5,8))

sns.violinplot(data=df, y='parental level of education', x='math score',
            hue='test preparation course', palette='Set2', split=True, inner=None)  # passing in inner param as None

plt.legend(bbox_to_anchor=(1.5, 0.5))
```

```python
# inner param ---> quartile option, draws line on those quartile on violin plot 

plt.figure(figsize=(5,8))

sns.violinplot(data=df, y='parental level of education', x='math score',
            hue='test preparation course', palette='Set2', split=True, inner='quartile')  # passing in inner param as quartile

plt.legend(bbox_to_anchor=(1.5, 0.5))
```

--> we can see quartile drwn for each KDE on violin plot showing us lot of info

---> recommended if we do quartile on violin plot that we dont do sub categoristion base of hue etc


```python

plt.figure(figsize=(5,8))

sns.violinplot(data=df, y='parental level of education', x='math score',
                inner='quartile')  # passing in inner param as quartile
```

---> we can see the clear distribution  & quartile if audience is already familiar with violin plot

```python
# inner param - stick option, drwas line for every instance showing distribution along with kde

plt.figure(figsize=(5,8))

sns.violinplot(data=df, y='parental level of education', x='math score',
            inner='stick')  # passing in inner param as stick
```

---> we can see line for every instance along with kde 

```python
# bandwidth, bw argument in violin plot 

#---> bandwidht in kde plot --- how much noise we want pick up when stacking & summing together those gaussian distrubution

plt.figure(figsize=(5,8))

sns.violinplot(data=df, y='parental level of education', x='math score', bw = 0.1)  # passing in bandwiwth bw param as 0.1

```

----> we can see kde pick up more noise as we pass in bw to more smaller value

----> i.e, the more we increase the bw valu , less noise it picks up

----> * report the default kde plot as shown by seaborn ( dont edit bandwidth ) 


#NB - Also we can see the main differences in distribution actually based of mini boxplot here 


----> i.e it is recoommend to just use Boxplot to shoow distribution within categories *But we can create interting plot using violin plot as we can see

### Swarm Plot

```python
# swarm plot ---> shows dots for all the entries

sns.swarmplot(data=df, x='math score');  # swarm plot distribution for single continuos feature, eg - math score   
```

```python
#--> we can see warning that some pts cannot be placed, i.e we have to either increase figure size or lower the dot size 
    #todisplay all the ponts

#--> generally we can just change the size of dots
plt.figure(figsize=(10,5))

sns.swarmplot(data=df, x='math score', size=2) # lower size by 2
```

----> it shows a mix between all the pts vs a violin plot, i.e we see kde distribution in terms of actual data pts

----> for large dataset, it ll have too many pts, we need to aggregate with boxplot or violin plot

```python
# seperate swarm plot base of categorical column

plt.figure(figsize=(8,5))

sns.swarmplot(data=df, y= 'gender', size=2, x='math score') # ---> seprating base of gender category 
```

```python
# hue param in swarm plot

plt.figure(figsize=(12,6))

sns.swarmplot(data=df, y= 'gender', size=2, x='math score', hue='test preparation course')

plt.legend(bbox_to_anchor=(1.25, 0.5))
```

---> we can see the coloring of pts v/s some who did complete test prep v/s someone not completing test prep which really hard
to read

---> * hue is not informative with swarmplot, instead what we probaly want to do is pass along dodge param

```python
# dodge param in swarm plot ---> split up wsarm plot for that secondary hue categoty we ask for

plt.figure(figsize=(12,6))

sns.swarmplot(data=df, y= 'gender', size=2, x='math score', hue='test preparation course', dodge=True)

plt.legend(bbox_to_anchor=(1.25, 0.5))
```

----> this is all it is to know about swarm plots

### boxen plot

```python
# boxen plot --> it is treated same as boxplot but with more infos & has all the params available in boxplot

plt.figure(figsize=(12,6))

sns.boxenplot(data=df, x = 'math score', y='test preparation course') # boxen plot for math score distribution against test prep
```

```python
# hue in boxen plot

plt.figure(figsize=(12,6))

sns.boxenplot(data=df, x = 'math score', y='test preparation course', hue='gender')
```

----> contains more infos but the audience need to be educated to understand what boxen plot type is

----> recommede we compare boxplot v/s boxen plot  to see difference & we will see that we get better understanding overall distribution in boxen plot compare to standard boxplot
